Q1. Explain Class and Object in OOP. Give example.

Class
A class is a blueprint (design) for creating objects.
It defines properties (data) and methods (functions).

Object
An object is a real instance of a class.

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Ravi", 20)   # object
```

Q3. Why is the __init__() function used? Give example.

The __init__() function is a constructor.
It is used to initialize (set) data when an object is created.

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

c1 = Car("Toyota")
```

Q5. What is inheritance? Give an example for each type of inheritance.

Inheritance means one class (child) gets properties and methods from another class (parent).

1️⃣ Single Inheritance

```python
class A:
    pass

class B(A):
    pass
```

Multiple Inheritance

```python
class A:
    pass

class B:
    pass

class C(A, B):
    pass
```

Multilevel Inheritance

```python
class A:
    pass

class B(A):
    pass

class C(B):
    pass
```

Hierarchical Inheritance

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass
```

Hybrid Inheritance

It is a combination of two or more types of inheritance.

```python
class A:
    pass

class B(A):
    pass

class C(A):
    pass

class D(B, C):
    pass
```
